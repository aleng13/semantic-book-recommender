```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata") #Downloading the kaggle dataset

print("Path to dataset files:", path) #Stored the dataset into a path
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv") #To read our books.csv file(essentially to see all the contents stored in it)
```

```python
books
```

```python
import seaborn as sns                 # For pretty visualizations like heatmaps
import matplotlib.pyplot as plt       # For creating and customizing plots

# Create a matplotlib axes object (where the plot will be drawn)
ax = plt.axes()

# Plot a heatmap of missing values in the dataset (rotated for better visibility)
# .isna() returns a boolean DataFrame of missing values
# .transpose() flips rows and columns to visualize columns as rows
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

# Set axis labels
plt.xlabel("Rows (Book Entries)")
plt.ylabel("Columns (Features with Missing Values)")

# Show the heatmap
plt.show()
```

```python
import numpy as np  # For numerical operations and array-based logic

# Create a new column that marks whether the 'description' is missing (1) or not (0)
books["missing_description"] = np.where(books["description"].isna(), 1, 0)

# Create a new column that estimates how old the book is (2024 minus published year)
books["age_of_book"] = 2024 - books["published_year"]

# Select columns we want to analyze correlation for
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# Create a correlation matrix using the Spearman method
# Spearman is used when relationships might be non-linear
correlation_matrix = books[columns_of_interest].corr(method="spearman")

# Set the theme for Seaborn plots
sns.set_theme(style="white")

# Set the figure size
plt.figure(figsize=(8, 6))

# Draw the correlation heatmap
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,             # Show values in each cell
    fmt=".2f",              # Format the numbers to 2 decimal places
    cmap="coolwarm",        # Set color scheme (blue to red)
    cbar_kws={"label": "Spearman Correlation"}  # Add label to color bar
)

# Set the title of the plot
heatmap.set_title("Correlation Heatmap")

# Show the plot
plt.show()
```

```python
#Display all the books having all or any of these fields missing so that we can delete them

books[ (books["description"].isna()) | (books["num_pages"].isna()) |
       (books["average_rating"].isna()) | (books["published_year"].isna()) ]

#The result shows that there around 303 books with missing details which is not that significant.
```

```python
#All the books that have the details

book_missing = books[ ~(books["description"].isna()) & ~(books["num_pages"].isna()) &
       ~(books["average_rating"].isna()) & ~(books["published_year"].isna()) ]
```

```python
#Even the descriptions are not that accurate or properly describing the book
#calculating: how many words are in each book’s description
#.str.split() → splits the description into a list of words
#.str.len() → counts the number of words in each list

book_missing = books[books["description"].notna()].copy()

book_missing["words_in_description"] = book_missing["description"].str.split().str.len()

#As you can see the distribution of the words in the description are not even
```

```python
# trying to fillter all the books having description words >= 25
#There are around 5000+ books
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
'''You're creating a new column called title_and_subtitle where:
If a book doesn’t have a subtitle, just use the title.
If a book does have a subtitle, combine the title and subtitle like this:
Title:Subtitle'''

'''Some books don’t have subtitles → you just want the title
But if subtitles exist → you want to preserve and merge them cleanly
Makes it easier to display or recommend books with a full name'''

book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25].copy()


book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(":".join, axis=1)))
```

```python
book_missing_25_words
```

```python
#Trying to create a new description column attached with a unique identifier
#created a new column tagged_description, attaching isbn with the description


'''If two books have the same description (e.g. “A thrilling mystery novel”), they’ll still be distinguishable by the ISBN
    If you’re going to feed this into:
A transformer model
Sentence embeddings
LangChain or Chroma vector store
→ Adding isbn13 gives context and keeps track of which book is which after processing.'''

book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```
